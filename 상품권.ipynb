```python
import pandas as pd
PATH = './sample_data/'
FILE = '온누리상품권.csv'
file = PATH+FILE
df = pd.read_csv(file, encoding = 'cp949') #cp949: 윈도우OS 기본 인코딩 포맷
df.head(3)
```

```python
df.columns
```

```python
# df.drop( '2013년',axis=1, inplace=True)
# df.drop( '2014년',axis=1, inplace=True)
df.drop( '2015년 1분기',axis=1, inplace=True)
df.drop( '2015년 2분기',axis=1, inplace=True)
df.drop( '2015년 3분기',axis=1, inplace=True)
df.drop( '2015년 4분기',axis=1, inplace=True)
df.drop( '2016년 1분기',axis=1, inplace=True)
df.drop( '2016년 2분기',axis=1, inplace=True)
df.drop( '2016년 3분기',axis=1, inplace=True)

df.drop( '2016년 4분기',axis=1, inplace=True)
df.drop( '2017년 1분기',axis=1, inplace=True)
df.drop( '2017년 2분기',axis=1, inplace=True)
df.drop( '2017년 3분기',axis=1, inplace=True)
df.drop( '2017년 4분기',axis=1, inplace=True)
df.drop( '2018년 1분기',axis=1, inplace=True)
df.drop( '2018년 2분기',axis=1, inplace=True)
df.drop( '2018년 3분기',axis=1, inplace=True)
df.drop( '2018년 4분기',axis=1, inplace=True)

df

```

```python
df.columns
```

```python
df.columns = ['지역명', '2010년', '2011년', '2012년', '2013년', '2014년']
df.columns
```

```python
df[df['지역명'].str.contains('제주')]

df
```

```python
df_total_group=df[df['지역명'].str.contains('제주')]
df_total_group
```

```python
df_total_group.columns
```

```python
import matplotlib.pyplot as plt

# 코랩에서 한글 폰트 설치한 후 사용하기!!!
# plt.rc('font', family='cmb10')                  # (코랩)한글 폰트
plt.rc('font', family='NanumBarunGothic')       # (코랩)한글 폰트
# plt.rc('font', family='Malgun Gothic')          # (win)한글 폰트


# 1.그래프 데이터 지정하기(X축, Y축)
x = df_total_group.index    # 지역명
y = df_total_group['지역명']   # 인구수

# 2.그래프 그리기
plt.bar(x, y,color='k')

# 3. 그래프 꾸미기
# plt.title('대한민국 "구씨"를 찾아라') # 그래프 제목
plt.title('Find Familyname Goo!!! ') # 그래프 제목
plt.xlabel('지역')                    # x축 이름
plt.ylabel('인구수')                  # y축 이름
plt.xticks(rotation = 45)             # X축 레이블 회전
plt.show()                            # 그래프 출력하기
```
